```python
from csv import reader
lst1 = []
with open('data/Migration1.csv', 'r') as csvfile:
    for line in reader(csvfile):
        lst1.append(line)
lst1 = lst1 = lst1[1:len(lst1)-5] 
for i in range(len(lst1)):
    for j in range(6, len(lst1[0])):    
        if lst1[i][j] == '':
            lst1[i][j] = 0
        else:
            lst1[i][j] = int(lst1[i][j])

clean1= []
year = 1960
for i in range(6,len(lst1[0])):
    for j in range(len(lst1)):
        temp = []
        for z in range(6):
            temp.append(lst1[j][z])
        temp.append(year)
        temp.append(lst1[j][i])
        clean1.append(temp)
    year += 10
clean1
```

```python
lst2 = []
with open('data/Population.csv', 'r') as csvfile:
    for line in reader(csvfile):
        lst2.append(line)
lst2 = lst2[1:]
for i in range(len(lst2)):
    for j in reversed(range(4, len(lst2[0]))):    
        if lst2[i][j] == '..':
            lst2[i][j] = lst2[i][j+1]
        else:
            lst2[i][j] = int(lst2[i][j])
lst2
```

```python
clean2 = []
for i in range(len(clean1)):
    for j in range(len(lst2)):
        if clean1[i][1] == lst2[j][3]:
            temp = []
            for z in range(len(clean1[i])):
                temp.append(clean1[i][z])
            
            if clean1[i][6] == 1960:
                temp.append(lst2[j][4])
            elif clean1[i][6] == 1970:
                temp.append(lst2[j][5])
            elif clean1[i][6] == 1980:
                temp.append(lst2[j][6])
            elif clean1[i][6] == 1990:
                temp.append(lst2[j][7])
            elif clean1[i][6] == 2000:
                temp.append(lst2[j][8])
            else:
                temp.append(0)
            clean2.append(temp)

clean2
```

```python
count = 0
country_lst = []
for i in range(len(clean2)):
    if len(clean2[i]) == 9:
        if clean2[i][0] not in country_lst:
            country_lst.append(clean2[i][0])   
        count+=1      
print(count)
print(country_lst)
len(country_lst)
```

```python
title =['Country', 'InorOut','Develop', 1960,1970,1980,1990,2000]
parallel_lst = []
parallel_lst.append(title)
from_country = []
to_country =[]
for i in range(len(lst1)):
    if lst1[i][0] not in from_country:
        from_country.append(lst1[i][0])
    if lst1[i][4] not in to_country:
        to_country.append(lst1[i][4])
#print(len(from_country))
#print(len(to_country))
developed_country = ['Andorra', 'Faroe Islands', 'Ireland', 
                     'Monaco', 'Spain', 'Australia', 'Finland', 
                     'Israel', 'Netherlands', 'Sweden', 'Austria', 
                     'France', 'Italy', 'New Zealand', ' Switzerland', 
                     'Belgium', 'Germany', 'Japan', 'Norway', 'Turkey', 
                     'Bermuda', 'Greece', 'Liechtenstein', 'Portugal', 
                     'United Kingdom' ,'Canada', 'Holy See', 'Luxembourg', 
                     'San Marino', 'United States' ,'Denmark', 'Iceland']
print(developed_country )
for j in country_lst:
    templst = [j]
    count60 = 0
    count70 = 0
    count80 = 0
    count90 = 0
    count00 = 0
    if j in developed_country:
        develop = 'Developed'
    else:
        develop = 'Developing'
    for i in range(len(lst1)):
        if (lst1[i][0] == j) and (lst1[i][4] in country_lst) :
            count60 += lst1[i][6]
            count70 += lst1[i][7]
            count80 += lst1[i][8]
            count90 += lst1[i][9]
            count00 += lst1[i][10]
    templst.append('Moveout')
    templst.append(develop)
    templst.append(count60)
    templst.append(count70)
    templst.append(count80)
    templst.append(count90)
    templst.append(count00)
    parallel_lst.append(templst)
    
for j in country_lst:
    templst = [j]
    count60 = 0
    count70 = 0
    count80 = 0
    count90 = 0
    count00 = 0
    if j in developed_country:
        develop = 'Developed'
    else:
        develop = 'Developing'
    for i in range(len(lst1)):
        if (lst1[i][4] == j) and (lst1[i][0] in country_lst):
            count60 += lst1[i][6]
            count70 += lst1[i][7]
            count80 += lst1[i][8]
            count90 += lst1[i][9]
            count00 += lst1[i][10]
    templst.append('Movein')
    templst.append(develop)
    templst.append(count60)
    templst.append(count70)
    templst.append(count80)
    templst.append(count90)
    templst.append(count00)
    parallel_lst.append(templst)    
print(parallel_lst)            
        



```
